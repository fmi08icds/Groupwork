# Spotify Dataset Description

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datasource as ds
```

Download the Spotify Track Dataset here: https://www.kaggle.com/datasets/zaheenhamidani/ultimate-spotify-tracks-db

Unzip and store the csv-file in the same directory or change variable path to your path.

```python
path = './SpotifyFeatures.csv'
```

load_spotify_dataset(path) will load the original dataset from the SpotifyFeatures.csv into data.

```python
data = ds.load_spotify_dataset(path=path)
data.head(10)
```

load_X_y(path, attribute_list, sample_size) extracts the feature matrix X and the vector y as numpy arrays. 

- Path: your path to SpotifyFeatures.csv
- attribute_list:  Choose attributes from ['popularity', 'acousticness', 'danceability', 'duration_ms', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']
- sample_size: 100 as default 


```python
X,y = ds.load_X_y(path=path)
print(X)
print(y)
```

```python
attribute_list = ['popularity', 
 'acousticness',
 'danceability',
 'duration_ms',
 'energy',
 'instrumentalness',
 'liveness',
 'loudness',
 'speechiness',
 'tempo',
 'valence']

for attribute in attribute_list:

    plt.hist(data[attribute], bins=20, color='lightblue',edgecolor='black')
    plt.xlabel('Index')
    plt.ylabel('# Spotify Tracks')
    plt.title(attribute)
    plt.show()
```
